# PaddleOCR + TrOCR

```python
from paddleocr import PaddleOCR
import numpy as np
from PIL import Image
from transformers import TrOCRProcessor, VisionEncoderDecoderModel
import torch
import re

# ‚úÖ PaddleOCR ch·ªâ d√πng ƒë·ªÉ detect bbox
paddleocr = PaddleOCR(lang="vi")

# ‚úÖ Model TrOCR d√†nh cho text in
TROCR_MODEL = "microsoft/trocr-base-printed"

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", device)

# ‚úÖ Load TrOCR ‚Äì th√™m trust_remote_code n·∫øu version m·ªõi
processor = TrOCRProcessor.from_pretrained(
    TROCR_MODEL,
    trust_remote_code=True
)

model = VisionEncoderDecoderModel.from_pretrained(
    TROCR_MODEL,
    trust_remote_code=True
).to(device)

print("‚úÖ TrOCR loaded th√†nh c√¥ng")
```

```python

def detect_text_boxes(image_path):
    """
    Nh·∫≠n ƒë·∫ßu v√†o: path ·∫£nh
    Tr·∫£ v·ªÅ: list bbox d·∫°ng [x1, y1, x2, y2]
    """

    # Ch·∫°y detect
    result = paddleocr.predict(image_path)

    bboxes = []

    for box in result[0]['rec_boxes']:
        bboxes.append(box)

    return bboxes


```

```python
image_path = "doc.png"  # ƒë·ªïi th√†nh file b·∫°n c√≥
bboxes = detect_text_boxes(image_path)

print("S·ªë v√πng ph√°t hi·ªán:", len(bboxes))
for b in bboxes:
    print(b)
```

```python
from PIL import Image

def crop_boxes(image_path, bboxes):
    """
    Input:
        image_path: ƒë∆∞·ªùng d·∫´n ·∫£nh g·ªëc
        bboxes: list bbox d·∫°ng [x1, y1, x2, y2]
    Output:
        list c√°c ·∫£nh crop (PIL.Image)
    """
    img = Image.open(image_path).convert("RGB")
    crops = []

    for (x1, y1, x2, y2) in bboxes:
        crop = img.crop((x1, y1, x2, y2))
        crops.append(crop)

    return crops
```

```python
crops = crop_boxes(image_path, bboxes)

print("S·ªë ·∫£nh crop:", len(crops))
crops[0].show()   # hi·ªÉn th·ªã crop d√≤ng ƒë·∫ßu
```

```python
def normalize_case(text):
    text = text.lower()  # to√†n b·ªô lower
    text = re.sub(r'([.!?]\s+)([a-z])', lambda m: m.group(1) + m.group(2).upper(), text)  # vi·∫øt hoa ƒë·∫ßu c√¢u
    if text: text = text[0].upper() + text[1:]
    return text

def trocr_read(image):
    # image: PIL.Image crop
    
    inputs = processor(image, return_tensors="pt").pixel_values.to(device)

    with torch.no_grad():
        generated_ids = model.generate(
            inputs,
            max_new_tokens=256,
            num_beams=4,
            early_stopping=True
        )

    text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0].strip()
    return normalize_case(text)
```

```python
texts = [trocr_read(c) for c in crops]

for t in texts:
    print(t)
```

```python
paragraph = "\n".join(texts)
print("\n===== OCR Paragraph =====\n")
print(paragraph)
```

```python
def summarize_text(text):
    """
    S·ª≠ d·ª•ng model summaries ƒë·ªÉ t√≥m t·∫Øt vƒÉn b·∫£n.
    Model: mT5, ViT5,...
    """
    inputs = tokenizer(
        f'T√≥m t·∫Øt: {text}',
        return_tensors="pt",
        truncation=True,
        max_length=512
    ).to(device)

    outputs = model.generate(
        **inputs,
        max_length=250,
        num_beams=5,
        early_stopping=True,
    )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
import os, unicodedata, shutil
from collections import Counter
from tqdm import tqdm
from PIL import Image

# C√°c th∆∞ m·ª•c rec/
SRC_FOLDERS = [
    r"D:\Study\Projects\HK7\QuichQuiz\archive\VietOCR-Paddle\InkData_line_processed\rec",
    r"D:\Study\Projects\HK7\QuichQuiz\archive\VietOCR-Paddle\vi_00\rec",
    r"D:\Study\Projects\HK7\QuichQuiz\archive\VietOCR-Paddle\vi_01\rec",
    r"D:\Study\Projects\HK7\QuichQuiz\archive\VietOCR-Paddle\meta\rec",
    r"D:\Study\Projects\HK7\QuichQuiz\archive\VietOCR-Paddle\random\rec",
]

OUT_DIR = r"D:\Study\Projects\HK7\QuichQuiz\dataset"
OUT_IMG_DIR = os.path.join(OUT_DIR, "images")
OUT_LABEL_PATH = os.path.join(OUT_DIR, "labels.txt")

SUBDIR_IMAGES = "train"
LABEL_FILENAME = "rec_gt_train.txt"

os.makedirs(OUT_IMG_DIR, exist_ok=True)

def normalize(text):
    return unicodedata.normalize("NFC", text).strip()

def parse_label(line):
    # Label format:  train/abc.png text
    parts = line.strip().split(maxsplit=1)  # t√°ch theo m·ªçi whitespace: space, tab...
    if len(parts) < 2:
        return None, None
    return parts[0].strip(), parts[1].strip()

def subset_name(folder):
    return os.path.basename(os.path.dirname(folder))
```

```python
subset_reports = []

for folder in SRC_FOLDERS:
    name = subset_name(folder)
    img_dir = os.path.join(folder, SUBDIR_IMAGES)
    label_file = os.path.join(folder, LABEL_FILENAME)

    print(f"\nüîé CHECK DATASET: {name}")

    if not os.path.exists(img_dir) or not os.path.exists(label_file):
        print("‚ö†Ô∏è Thi·∫øu th∆∞ m·ª•c ·∫£nh ho·∫∑c file nh√£n ‚Üí b·ªè qua")
        continue

    images = [f for f in os.listdir(img_dir) if f.lower().endswith(("png","jpg","jpeg"))]
    labels_raw = open(label_file, encoding="utf8", errors="ignore").read().splitlines()

    label_items, bad = [], 0
    for ln in labels_raw:
        fn, txt = parse_label(ln.strip())
        if not fn or not txt:
            bad += 1
        else:
            label_items.append((fn, normalize(txt)))

    img_set = set(images)
    label_set = {fn for fn, _ in label_items}

    missing_label = img_set - label_set
    missing_img   = label_set - img_set
    dups = [k for k,v in Counter(fn for fn,_ in label_items).items() if v > 1]

    print(f"üìÅ ·∫¢nh: {len(images)} | üìù D√≤ng nh√£n: {len(labels_raw)} | ‚úÖ Nh√£n h·ª£p l·ªá: {len(label_set)}")
    print(f"‚ùå Nh√£n l·ªói d√≤ng: {bad}")
    print(f"‚ö†Ô∏è ·∫¢nh thi·∫øu nh√£n: {len(missing_label)}")
    print(f"‚ö†Ô∏è Nh√£n thi·∫øu ·∫£nh: {len(missing_img)}")
    print(f"‚ôªÔ∏è File tr√πng trong nh√£n: {len(dups)}")

    subset_reports.append({
        "name": name,
        "img_dir": img_dir,
        "label_items": label_items,
        "img_set": img_set,
        "label_set": label_set,
        "num_img": len(images)
    })
```

```python
used = set()
total = 0

with open(OUT_LABEL_PATH, "w", encoding="utf8") as label_out:
    for r in subset_reports:
        print(f"\nüì¶ G·ªôp subset: {r['name']} ({r['num_img']} ·∫£nh)")

        for fn, text in tqdm(r["label_items"], desc=f"Merging {r['name']}"):
            img_name = os.path.basename(fn.strip())  # l·∫•y t√™n file cu·ªëi
            src = os.path.join(r["img_dir"], img_name)
            if not os.path.exists(src) or not text:
                print("‚ùóFILE NOT FOUND:", src)
                continue

            if not os.path.exists(src) or not text:
                continue

            # tr√°nh ƒë√® file
            out_name = img_name
            if out_name in used:
                out_name = f"{r['name']}__{img_name}"

            dst = os.path.join(OUT_IMG_DIR, out_name)
            shutil.copy(src, dst)

            label_out.write(f"images/{out_name}|{text}\n")
            used.add(out_name)
            total += 1

print(f"\n‚úÖ DONE MERGING")
print(f"üñºÔ∏è T·ªïng ·∫£nh ƒë√£ g·ªôp: {total}")
print(f"üìÑ labels.txt: {OUT_LABEL_PATH}")
print(f"üìÇ images/: {OUT_IMG_DIR}")
```

```python
import os, random, shutil
from tqdm.notebook import tqdm

# ==== CONFIG ====
LABEL_FILE = "dataset/labels.txt"
IMAGE_FOLDER = "dataset"
OUT_DIR = "data"

TRAIN_RATIO, VAL_RATIO, TEST_RATIO = 0.7, 0.1, 0.2
# =================

print("üîç ƒê·ªçc file nh√£n...")
with open(LABEL_FILE, "r", encoding="utf-8") as f:
    lines = f.readlines()

print(f"üìÑ T·ªïng s·ªë d√≤ng trong labels.txt: {len(lines)}")

# Shuffle ƒë·ªÉ tr√°nh bias
random.shuffle(lines)

# T√≠nh s·ªë m·∫´u
total = len(lines)
train_size = int(total * TRAIN_RATIO)
val_size = int(total * VAL_RATIO)

train_data   = lines[:train_size]
val_data     = lines[train_size:train_size + val_size]
test_data    = lines[train_size + val_size:]

print(f"\nüìä Chia d·ªØ li·ªáu:")
print(f" - Train: {len(train_data)}")
print(f" - Valid: {len(val_data)}")
print(f" - Test : {len(test_data)}")

# ===== T·∫°o th∆∞ m·ª•c =====
for split in ["train", "valid", "test"]:
    os.makedirs(f"{OUT_DIR}/{split}/images", exist_ok=True)

# ===== Ghi labels =====
print("\nüìù L∆∞u file labels...")
open(f"{OUT_DIR}/train/labels.txt", "w", encoding="utf-8").writelines(train_data)
open(f"{OUT_DIR}/valid/labels.txt", "w", encoding="utf-8").writelines(val_data)
open(f"{OUT_DIR}/test/labels.txt" , "w", encoding="utf-8").writelines(test_data)

# ===== H√†m copy ·∫£nh =====
def copy_images(data, split_name):
    missing = 0
    print(f"\nüìÅ Copy ·∫£nh ‚Üí {split_name} ...")

    for line in tqdm(data):
        raw_name = line.split("|")[0].strip()

        # Ch·ªâ l·∫•y t√™n file, b·ªè prefix "images/"
        img_name = os.path.basename(raw_name)

        # ƒê∆∞·ªùng d·∫´n g·ªëc
        src = os.path.join(IMAGE_FOLDER, raw_name)
        
        # T·∫°o ƒë√∫ng ƒë∆∞·ªùng d·∫´n ƒë√≠ch
        dst = os.path.join(f"{OUT_DIR}/{split_name}/images", img_name)

        print(f"SRC: {src}  ‚Üí DST: {dst}")

        # Ki·ªÉm tra file
        if not os.path.exists(src):
            missing += 1
            print(f"‚ö†Ô∏è Thi·∫øu ·∫£nh: {src}")
            continue

        shutil.copy(src, dst)

    print(f"‚úÖ Done {split_name} ‚Äî thi·∫øu {missing} ·∫£nh\n")



# ===== Copy ·∫£nh =====
copy_images(train_data, "train")
copy_images(val_data, "valid")
copy_images(test_data, "test")

print("üéØ Ho√†n t·∫•t t√°ch d·ªØ li·ªáu!")
```
